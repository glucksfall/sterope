```python
import glob
import numpy
import pandas
import seaborn
import matplotlib.pyplot as plt
```

```python
with open('./results_1551219370/reports/report_DINfluxes_S1.txt') as file:
    S1 = pandas.read_csv(file, delimiter='\t')
    
with open('./results_1551219370/reports/report_DINfluxes_S1_conf.txt') as file:
    S1_conf = pandas.read_csv(file, delimiter='\t')
```

```python
fig, ax = plt.subplots(1, 1, figsize = (4, 3), dpi = 200)
seaborn.barplot(y = S1[S1.columns[2:]].columns, x = S1.iloc[0, 2:], ax = ax, **{ 'xerr' : S1_conf.iloc[0, 2:]})

ax.set_xlabel(f'Influence {S1.iloc[0,0]} over {S1.iloc[0,1]}')

seaborn.despine()
plt.tight_layout()
plt.show()
```

```python
with open('./results_1551219370/reports/report_DINfluxes_S1.txt') as file:
    S1 = pandas.read_csv(file, delimiter='\t', index_col = [0,1])
    
with open('./results_1551219370/reports/report_DINfluxes_S1_conf.txt') as file:
    S1_conf = pandas.read_csv(file, delimiter='\t', index_col = [0,1])
```

```python
for rule_1st, rule_2nd in zip(S1.index.levels[0], S1.index.levels[1])
    fig, ax = plt.subplots(1, 1, figsize = (4, 3), dpi = 200)
    seaborn.barplot(y = S1.columns, x = S1.loc[rule_1st].loc[rule_2nd, :], ax = ax, 
                    **{ 'xerr' : S1_conf.loc['r1'].loc['r1', :] })

    if rule_1st == rule_2nd:
        ax.set_xlabel(f'Influence {rule_1st} over itself')
    else:
        ax.set_xlabel(f'Influence {rule_1st} over {rule_2nd}')

    seaborn.despine()
    plt.tight_layout()
    plt.show()
```
