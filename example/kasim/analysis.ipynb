```python
din_hits = []
din_fluxes = []

for flux in files:
    with open(flux, 'r') as file:
        data = pandas.read_json(file)
    
    din_hits.append(data['din_hits'].iloc[1:].values)
    din_fluxes.append(pandas.DataFrame([x for x in data['din_fluxs']]).values)
```
